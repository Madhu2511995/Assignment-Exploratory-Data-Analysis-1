1. Mean/Median Imputation:
-----------------------------------
#### Advantages:
- Easy to implement.
- Preserves the overall distribution of the feature.
#### Disadvantages:
- May introduce bias, especially if the data is not missing completely at random (MCAR).
- Reduces the variability in the data, potentially underestimating the uncertainty.

2. Mode Imputation (for categorical data):
---------------------------------------------------
#### Advantages:
- Suitable for categorical variables.
- Maintains the most frequent category.
#### Disadvantages:
- Similar to mean/median imputation, it may introduce bias.
- May not be appropriate for nominal data with no natural ordering.

3. Removing Rows with Missing Data (Listwise Deletion):
-------------------------------------------------------------
#### Advantages:
- Simple and straightforward.
- Preserves the original data structure.

#### Disadvantages:
- Loss of potentially valuable information, especially if missing data is not completely random.
- Can reduce the sample size, potentially affecting model performance and representativeness.

### Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the Wine Quality dataset

# Separate features (X) and the target variable (y)
X = df.drop('quality', axis=1)

# Standardize the feature matrix (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA()
pca.fit(X_scaled)

# Calculate the cumulative explained variance
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = explained_variance_ratio.cumsum()

# Determine the minimum number of components to explain 90% of the variance
n_components_90 = (cumulative_variance >= 0.90).sum() + 1  # Add 1 to start from 1 component

# Plot the explained variance ratio
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(explained_variance_ratio) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Explained Variance Ratio')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()

print(f"Minimum number of components to explain 90% of the variance: {n_components_90}")
```
